```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_tracks=pd.read_csv("tracks.csv")
```

```python
df_tracks.head()
```

```python
# to find count of null values in each col 
pd.isnull(df_tracks).sum() 
```

```python
df_tracks.info()
```

```python
pop=df_tracks.sort_values("popularity",ascending=False)
pop=pop.set_index([pd.Index(range(1,586673,1))])
pop.head()
```

```python
# find name of artist most popular
pop.loc[1,"artists"]
```

```python
# top popular tracks >90%
top_popular=df_tracks[df_tracks.popularity>90].sort_values("popularity",ascending=False)
top_popular
```

```python
# make release date as index and convert its datatype to datetime
df_tracks.set_index("release_date",inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# convert tracks duration from ms to s
df_tracks["duration"]=df_tracks["duration_ms"].apply(lambda x: x/1000)
df_tracks.drop("duration_ms",inplace=True,axis=1)
df_tracks.head()
```

```python
# drop unneccessary columns
df_tracks.drop("explicit",inplace=True,axis=1)
df_tracks.drop("key",inplace=True,axis=1)
df_tracks.drop("mode",inplace=True,axis=1)
```

```python
# correlation heat map 
corr_df=df_tracks.corr(method="pearson")
heatmap=sns.heatmap(corr_df,annot=False)
heatmap.set_title("Correlation Heatmap between variables")
heatmap.set_xticklabels(heatmap.get_xticklabels())
```

```python
# regression plot between loudness and energy
sns.regplot(data=df_tracks,y="loudness",x="energy",ci=None,color='r').set_title("regression plot between loudness and energy")
```

```python
# regression plot between popularity and acousticness
sns.regplot(data=df_tracks,y="popularity",x="acousticness",ci=None).set_title("regression plot between popularity and acousticness")
```

```python
# in above plot there are too much data points so we take 3%sample from df_tracks
sns.regplot(data=df_tracks.sample(int(0.003*len(df_tracks))),y="popularity",x="acousticness",ci=None).set_title("regression plot between popularity and acousticness")
```

```python
# draw graph showing freq of songs vary with years
df_tracks["dates"]=df_tracks.index.get_level_values("release_date")
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,kind="hist").set(title="Number of songs per year")
```

```python
# variation of tracks duration with years
total_dur=df_tracks.duration
fig,ax=plt.subplots(figsize=(14,8))
fig=sns.barplot(x=years,y=total_dur,errwidth=False).set(title="years vs duration")
plt.xticks(rotation=90)
```
